# Análise dos Dados de Filmes Exibidos no Brasil de 2009 a 2017
Com base nos dados disponibilizados pelo Observatório Brasileiros do Cinema e Audiovisual
da ANCINE (https://oca.ancine.gov.br) vamos obter os dados, em arquivo Excel, dos filmes exibidos no Brasil

```python
# Importando as bibliotecas
import pandas as pd

# Criando o DataFrame com base no arquivo Excel
# Precisamos pular as 2 primeiras linhas
# e as ultimas 16 linhas 
df = pd.read_excel('2120_0.xlsx', skiprows=2, skipfooter=16)

# Exibindo os 5 primeiros registros do DataFrame
df.head()
```

```python
# Vamos verificar as 5 ultimas linhas
df.tail()
```

```python
# Olhando os tipos de dados
df.dtypes
```

```python
# Precisamos corrigir o tipo do campo 'Data de lancamento'
# Mas primeiro vamos verificar se não existem valores incorretos na coluna
df['Data de lançamento'].value_counts()
```

```python
# Vamos verificar alguns filmes onde a coluna 'Data de lançamento' == 'Relançamento'
df[df['Data de lançamento'] == 'Relançamento'].head()
```

## Decisão sobre os dados incorretos/faltantes/divergentes
Muitas vezes encontramos um conjunto de dados que precisa ser tratado antes de fazermos as análises.

Esse tratamento pode envolver obter novos dados, modificar os dados ou até mesmo excluí-los. Isso depende da situação
encontrada, do objeivo da análise e da relevância dos dados problemáticos.

Nesse caso específico, vamos analisar apenas os filmes novos, lançados a cada ano. Sendo assim, o relançamento de filmes
não nos interessa e iremos descartar os dados problemáticos.

```python
# Para remover os filmes onde a coluna 'Data de lançamento' == 'Relançamento'
# vamos criar um novo DataFrame que não contem esses filmes
# o método copy() força que seja criada um novo DataFrame
df_novo = df[df['Data de lançamento'] != 'Relançamento'].copy()

df_novo.head()
```

```python
# Vamos verificar se ainda existem valores incorretos na coluna
df_novo['Data de lançamento'].value_counts()
```

```python
# Convertendo a coluna 'Data de lançamento'
df_novo['Data de lançamento'] = pd.to_datetime(df_novo['Data de lançamento'])

df_novo.dtypes
```

```python
# Quais os anos tiveram maior quantidade de lançamentos?
df_novo['Data de lançamento'].dt.year.value_counts()
```

```python
%matplotlib inline

df_novo['Data de lançamento'].dt.year.value_counts().plot.bar()
```

```python
# Qual o filme de maior bilheteria?
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].max()]
```

```python
# E qual o de menor renda?
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].min()]
```
